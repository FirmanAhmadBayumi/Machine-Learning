```python
# Melakukan pengambilan data

url = "https://raw.githubusercontent.com/rahmantaufik27/dataset/main/study/student_score_pass.csv"
data = pd.read_csv(url)
data
```

### Splitting Data

```python
# Melakukan penghapusan data yang bersifat NaN

data.dropna(inplace = True)
data
```

```python
# Mengubah data di kolom pass_test not pass menjadi 0, dan pass menjadi 1 
data['pass_test'] = data['pass_test'].replace('not pass', 0)
data['pass_test'] = data['pass_test'].replace('pass', 1)

# Mengkonversi nilai pada kolom pass_test ke tipe data float 
data['pass_test'] = data['pass_test'].astype(float)
```

### Modelling

```python
# Pemilihan Feature
# Pada program ini saya akan menghapus data pass_test

data_training = data.drop(["npm"], axis = 1)
data_training
```

```python
x_dataLatih, x_dataUji, y_dataLatih, y_dataUji = tts(data_training.drop(["pass_test"], axis = 1), data_training["pass_test"], test_size=0.2, random_state=2)
```

Deskripsi penjelasan pada kode program diatas:
1. test_size digunakan untuk menentukan ukuran data uji sebagai proporsi dari keseluruhan dataset
2. test_size = 0.2, artinya 20% dari data uji yang akan digunakan sebagai data tes, sedangkan 80% sisanya digunakan sebagai data latih.
3. random_state adalah argumen opsional yang digunakan untuk menentukan seed atau biji secara acak yang digunakan dalam pembagian data.



```python
model = AC().fit(x_dataLatih)
model
```

```python
# Melakukan prediksi pada data latih

predict_data_latih = model.fit_predict(x_dataLatih)

predict_data_latih
```

```python
# Melakukan prediksi pada data uji

predict_data_uji = model.fit_predict(x_dataUji)
predict_data_uji
```

### Accuracy Score

```python
# Menghitung nilai akurasi pada data latih

hitAkurasi_dataLatih = accScore(y_dataLatih, predict_data_latih)
hitAkurasi_dataLatih
```

```python
# Menghitung nilai akurasi pada data Uji

hitAkurasi_dataUji = accScore(y_dataUji, predict_data_uji)
hitAkurasi_dataUji
```

### Bonus
Hitung nilai akurasi data lebih dari 0.5

```python
# Random Forest Classifier digunakan dengan jumlah 500 estimators (jumlah pohon keputusan yang akan dibangun) dengan random_state sebesar 3.

model2 = RFS(n_estimators = 500, random_state = 3)
model2
```

```python
# Melatih model Random Forest Classifier menggunakan data latih
model2.fit(x_dataLatih, y_dataLatih)

# Melakukan prediksi pada data uji
predict_data_uji2 = model2.predict(x_dataUji)
predict_data_uji2
```

```python
# Menghitung akurasi pada data uji
akurasi = accScore(y_dataUji, predict_data_uji2)

# Menghitung dan mengubah hasil nilai akurasi ke persen
hitAkurasi = akurasi * 100
print("Akurasi :", hitAkurasi, "%")
```
