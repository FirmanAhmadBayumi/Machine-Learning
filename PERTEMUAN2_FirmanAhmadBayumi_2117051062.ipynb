```python
import pandas as pd
import numpy as np

#import apriory dari library mlextend
from mlxtend.frequent_patterns import apriori, association_rules
```

## Data Set


*   Membuat dataset taransaksi sebagai dataframe
*   Cek dataset transaksi

```python
# membuat data tabular yang terdiri dari data taransaksi

df = pd.DataFrame(
    {
        "ID Transaksi" : pd.Series([1, 2, 3, 4, 5]),
        "Item" : pd.Series([["Susu", "Roti", "Biskuit", "Mentega"],\
                           ["Susu", "Biskuit", "Gula", "Kopi"], \
                           ["Gula", "Kopi", "Susu"], ["Kopi", "Biskuit", "Susu", "Gula"], \
                           ["Mentega", "Roti", "Kopi"]]),
    }
)
```

```python
df.info()
```

```python
df
```

```python
df = df.explode('Item')
df
```

```python
basket = (df.groupby(['ID Transaksi', 'Item'])['Item'].count()\
                                      .unstack().reset_index().fillna(0)\
                                      .set_index('ID Transaksi'))

basket.head()
```

## Algoritma Apriori
Ekseskusi algoritma apriori menggunakan library mlextend dengan fungsi apriori

### Apriori Tahap 1
Eliminasi terhadap itemset berdasrkan minimum support

```python
# Set item minimum support 60%

frequent_itemsets = apriori(basket, min_support = 0.6, use_colnames = True)

#print(list(frequent_itemsets))
frequent_itemsets
```

### Apriori Tahap 2
Perhitungan Confidence untuk menyatkan keterkaitan antara 1 barang yang dibeli dengan barang yang lain
contoh: a membeli biskuit sudah pasti membeli susu

```python
rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold=0.6)
rules[["antecedents", "consequents", "support", "confidence"]]
```

```python
#inisiasi dataset
#untuk dataset ini saya buat sendiri

dataset = [
    ['Odol', 'Sikat Gigi', 'Sabun Mandi', 'Shampo', 'Sabun Wajah'],
    ['Sikat Gigi', 'Odol', 'Conditioner', 'Listerine'],
    ['Sikat Gigi', 'Odol', 'Listerine'],
    ['Shampo', 'Conditioner', 'Sikat Gigi', 'Odol'],
    ['Sabun Mandi', 'Shampo', 'Conditioner']
]
```

```python
#fungsi untuk menghitung kandidat itemset

def hitung_kandidat(prev_itemset, n):
    kandidat = []
    tmp = len(prev_itemset)
    
    for i in range(tmp):
        for j in range(i+1, tmp):

            #mengambil itemset sebelumnya yang memiliki (n-2) item yang sama
            if prev_itemset[i][:n-2] == prev_itemset[j][:n-2]:
              
                #menggabungkan dua itemset menjadi satu itemset kandidat baru
                kandidat1 = list(set(prev_itemset[i] + prev_itemset[j]))
                kandidat1.sort()
                
                #menambahkan itemset kandidat ke dalam list kandidat
                if kandidat1 not in kandidat:
                    kandidat.append(kandidat1)
    
    return kandidat

#fungsi untuk menghitung support count dari setiap itemset dalam dataset
def hitung_support_count(itemset, dataset):
    total = 0
    for trans in dataset:
        if set(itemset).issubset(set(trans)):
            total += 1
    return total
```

```python
#fungsi untuk menghitung itemset yang memiliki support count >= minimal support

def hitung_frequent_itemset(itemsets, min_support, dataset):
    frequent_itemset = {}
    for itemset in itemsets:
        support_count = hitung_support_count(itemset, dataset)
        support = float(support_count) / len(dataset)
        if support >= min_support:
            frequent_itemset[tuple(itemset)] = support
    return frequent_itemset

#fungsi utama untuk mengeksekusi algoritma Apriori
def apriori(dataset, min_support):
    itemset = [[item] for item in set([item for trans in dataset for item in trans])]
    frequent_itemset = hitung_frequent_itemset(itemset, min_support, dataset)
    print_frequent_itemset(frequent_itemset)
    
    n = 2
    while len(frequent_itemset) > 0:
        itemset = list(frequent_itemset.keys())
        candidates = hitung_kandidat(itemset, n)
        frequent_itemset = hitung_frequent_itemset(candidates, min_support, dataset)
        print_frequent_itemset(frequent_itemset)
        n += 1

#fungsi untuk menampilkan frequent itemset
def print_frequent_itemset(frequent_itemset):
    for itemset, support in frequent_itemset.items():
        print('{} : {}'.format(list(itemset), round(support, 2)))
```

```python
#mencari algoritma apriori dengan minimum support 60%
apriori(dataset, 0.6)
```
